```python
import os
import pandas as pd
```

```python
PROJECT_DIR = os.path.abspath("..")

RESULT_PATH = os.path.join(
    PROJECT_DIR,
    "data",
    "results",
    "retrieval_results.csv"
)

retrieval = pd.read_csv(RESULT_PATH)

retrieval
```

```python
DATA_PATH = os.path.join(
    PROJECT_DIR,
    "data",
    "processed",
    "cases.csv"
)

cases = pd.read_csv(DATA_PATH)

cases.head()
```

```python
best_case = retrieval.iloc[0]

best_case
```

```python
solution = cases[
    cases["case_id"] == best_case["case_id"]
]

solution
```

```python
print("="*60)
print("HASIL REUSE SOLUTION")
print("="*60)

print("Case ID :", solution.iloc[0]["case_id"])
print("Nomor Putusan :", solution.iloc[0]["nomor_putusan"])
print("Nama Terdakwa :", solution.iloc[0]["nama_terdakwa"])
print("Pasal :", solution.iloc[0]["pasal"])
print("Amar Putusan :")
print(solution.iloc[0]["amar_putusan"])
```

```python
reuse = pd.DataFrame({

    "query_case":[best_case["case_id"]],

    "nomor_putusan":[solution.iloc[0]["nomor_putusan"]],

    "nama_terdakwa":[solution.iloc[0]["nama_terdakwa"]],

    "pasal":[solution.iloc[0]["pasal"]],

    "amar_putusan":[solution.iloc[0]["amar_putusan"]],

    "similarity":[best_case["similarity"]]

})

reuse
```

```python
SAVE_DIR = os.path.join(
    PROJECT_DIR,
    "data",
    "results"
)

os.makedirs(SAVE_DIR, exist_ok=True)

reuse.to_csv(
    os.path.join(
        SAVE_DIR,
        "reuse_solution.csv"
    ),
    index=False,
    encoding="utf-8-sig"
)

print("Reuse Solution berhasil disimpan.")
```
