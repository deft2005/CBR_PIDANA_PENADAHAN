```python
import os
import pandas as pd
import numpy as np
```

```python
PROJECT_DIR = os.path.abspath("..")

RESULT_PATH = os.path.join(
    PROJECT_DIR,
    "data",
    "results",
    "retrieval_results.csv"
)

retrieval = pd.read_csv(RESULT_PATH)

retrieval
```

```python
threshold = 0.50

retrieval["prediction"] = np.where(
    retrieval["similarity"] >= threshold,
    "Relevant",
    "Not Relevant"
)

retrieval
```

```python
total_case = len(retrieval)

relevant = (retrieval["prediction"] == "Relevant").sum()

not_relevant = (retrieval["prediction"] == "Not Relevant").sum()

print("="*50)
print("HASIL EVALUASI")
print("="*50)
print(f"Jumlah Retrieval : {total_case}")
print(f"Relevant         : {relevant}")
print(f"Not Relevant     : {not_relevant}")
```

```python
print("="*50)

print(f"Similarity Max  : {retrieval['similarity'].max():.4f}")
print(f"Similarity Min  : {retrieval['similarity'].min():.4f}")
print(f"Similarity Mean : {retrieval['similarity'].mean():.4f}")
print(f"Similarity Std  : {retrieval['similarity'].std():.4f}")
```

```python
EVAL_DIR = os.path.join(
    PROJECT_DIR,
    "data",
    "eval"
)

os.makedirs(EVAL_DIR, exist_ok=True)

retrieval.to_csv(
    os.path.join(
        EVAL_DIR,
        "evaluation_results.csv"
    ),
    index=False,
    encoding="utf-8-sig"
)

print("Evaluation berhasil disimpan.")

```

```python
summary = pd.DataFrame({

    "Jumlah Retrieval":[total_case],

    "Relevant":[relevant],

    "Not Relevant":[not_relevant],

    "Similarity Max":[retrieval["similarity"].max()],

    "Similarity Mean":[retrieval["similarity"].mean()]

})

summary
```

```python
summary.to_csv(

    os.path.join(
        EVAL_DIR,
        "evaluation_summary.csv"
    ),

    index=False,

    encoding="utf-8-sig"

)

print("Summary berhasil disimpan.")
```
