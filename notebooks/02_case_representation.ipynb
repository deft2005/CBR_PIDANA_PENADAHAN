```python
import re
import pandas as pd
from pathlib import Path
from tqdm import tqdm
```

```python
BASE_DIR = Path.cwd().parent

RAW_DIR = BASE_DIR / "data" / "raw"
PROCESSED_DIR = BASE_DIR / "data" / "processed"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

txt_files = sorted(RAW_DIR.glob("*.txt"))

print("Jumlah TXT :",len(txt_files))
```

```python
def extract_court(text):

    match = re.search(

        r"Pengadilan\s+Negeri\s+([A-Za-z ]+)",

        text,

        re.IGNORECASE

    )

    if match:

        return "Pengadilan Negeri " + match.group(1).strip()

    return "-"
```

```python
def extract_year(text):

    match = re.search(

        r"/(\d{4})/",

        text

    )

    if match:

        return match.group(1)

    return "-"
```

```python
def extract_defendant(text):

    patterns = [

        r"Nama lengkap\s*:\s*(.+)",

        r"Nama\s*:\s*(.+)",

        r"Terdakwa\s*:\s*(.+)"

    ]

    for pattern in patterns:

        match = re.search(pattern,text,re.IGNORECASE)

        if match:

            return match.group(1).strip()

    return "-"
```

```python
def extract_article(text):

    matches = re.findall(

        r"Pasal\s+[0-9A-Za-z\s,().]+",

        text,

        re.IGNORECASE

    )

    if matches:

        return "; ".join(matches)

    return "-"
```

```python
def extract_verdict(text):

    text_upper = text.upper()

    keywords = [

        "MENGADILI",

        "MENETAPKAN",

        "MEMUTUSKAN"

    ]

    for key in keywords:

        if key in text_upper:

            idx = text_upper.find(key)

            return text[idx:idx+3000]

    return "-"
```
