```python
import os
import re
from pathlib import Path

import pandas as pd
import pdfplumber
from tqdm import tqdm
```

```python
# ==========================
# KONFIGURASI FOLDER
# ==========================

BASE_DIR = Path.cwd().parent

DATASET_DIR = BASE_DIR / "dataset"
RAW_DIR = BASE_DIR / "data" / "raw"
LOG_DIR = BASE_DIR / "logs"

RAW_DIR.mkdir(parents=True, exist_ok=True)
LOG_DIR.mkdir(parents=True, exist_ok=True)

print("Project :", BASE_DIR)
print("Dataset :", DATASET_DIR)
print("Raw :", RAW_DIR)
```

```python
# ==========================
# MENCARI FILE PDF
# ==========================

pdf_files = sorted(DATASET_DIR.glob("*.pdf"))

print("Jumlah PDF :", len(pdf_files))

for pdf in pdf_files:
    print(pdf.name)
```

```python
# ==========================
# CLEANING TEXT
# ==========================

def clean_text(text):

    if text is None:
        return ""

    text = text.replace("\n", " ")

    text = re.sub(r"\s+", " ", text)

    text = re.sub(r"Halaman\s+\d+", " ", text)

    text = re.sub(r"Disclaimer.*", " ", text)

    text = text.strip()

    return text
```

```python
# ==========================
# PDF -> TXT
# ==========================

summary = []

for i, pdf_file in enumerate(tqdm(pdf_files), start=1):

    full_text = ""

    with pdfplumber.open(pdf_file) as pdf:

        for page in pdf.pages:

            text = page.extract_text()

            if text:
                full_text += text + "\n"

    full_text = clean_text(full_text)

    txt_name = f"case_{i:03}.txt"

    txt_path = RAW_DIR / txt_name

    with open(txt_path, "w", encoding="utf-8") as f:

        f.write(full_text)

    summary.append({

        "Case_ID": i,

        "PDF_File": pdf_file.name,

        "TXT_File": txt_name,

        "Characters": len(full_text),

        "Words": len(full_text.split())

    })

print("Selesai membuat TXT")
```

```python
summary_df = pd.DataFrame(summary)

summary_df.head()
```

```python
summary_path = RAW_DIR / "raw_summary.csv"

summary_df.to_csv(summary_path, index=False)

print(summary_path)
```
